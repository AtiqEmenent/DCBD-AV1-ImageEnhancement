```python
%%bash

# Source and destination directories
SRC="Datasets/test/Data6FullImage/target"
DEST="Datasets/test/Data6FullImageCropped/target"

# Create destination directory if it doesn't exist
mkdir -p "$DEST"


for img in Datasets/test/Data6FullImage/target/*.bmp; do
    [ -e "$img" ] || continue
    ffmpeg -i "$img" -vf "crop=700:300" "Datasets/test/Data6FullImageCropped/target/$(basename "$img")"
done
```

```python
%%bash

# Source and destination directories
SRC="Datasets/test/Data6FullImage/input"
DEST="Datasets/test/Data6FullImageCropped/input"

# Create destination directory if it doesn't exist
mkdir -p "$DEST"


for img in Datasets/test/Data6FullImage/input/*.bmp; do
    [ -e "$img" ] || continue
    ffmpeg -i "$img" -vf "crop=700:300" "Datasets/test/Data6FullImageCropped/input/$(basename "$img")"
done
```
